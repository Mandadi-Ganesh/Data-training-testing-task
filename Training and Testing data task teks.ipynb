```python
import warnings
warnings.filterwarnings('ignore')
for i in x.columns:
    sns.distplot(x[[f'{i}']])
    plt.title(f'{i}')
    plt.show()
```

```python
def handle_outliers(column):
    q1, q3 = x[column].quantile([0.25, 0.75]) 
    iqr_value = q3 - q1                        
    lower_bound = q1 - 1.5 * iqr_value         
    upper_bound = q3 + 1.5 * iqr_value         
    

    x[column] = np.where(
        x[column] > upper_bound, upper_bound,
        np.where(x[column] < lower_bound, lower_bound, x[column])
    )
for column_name in x.columns:
    handle_outliers(column_name)
```

```python

plt.figure(figsize=(12, 6))
plt.gcf().set_facecolor('lightgrey')
sns.boxplot(data=x, palette="Blues")
plt.title(" Box Plot ")
plt.xticks(rotation=90)
plt.show()
```
